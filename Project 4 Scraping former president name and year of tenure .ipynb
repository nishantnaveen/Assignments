# Scraping former indian president and their year of tenure

```python
import requests   
import bs4 
from bs4 import BeautifulSoup

link= requests.get('https://presidentofindia.nic.in/former-presidents.htm')
link
soup=BeautifulSoup(link.content)

soup
```

```python
president_name=[]
term=[]
for i in soup.find_all('div', class_="presidentListing"):
    
    president_name.append(i.text.split('\n')[1].split('(')[0])
    term.append(i.text.split('\n')[2].split(':')[1])
print(f'The name of all the respected former presidents are {president_name},"\n"')
print(f'The term of office respectively are {term}')
# term=[]
# for i in soup.find_all('span', class_="terms"):
#     term.append(i.text.replace('\n', ""))
# term
```
